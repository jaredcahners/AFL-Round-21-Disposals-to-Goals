## Whose disposals lead to the most points?

```python
#import pandas
import pandas as pd
```

```python
#read the csv
r21 = pd.read_csv('AFL Play by play thru round 21.csv')
```

```python
r21.tail(20)
```

```python
#create a name column from first name and last name

r21['Player'] = r21['First_name'] + ' ' + r21['Last_name']
```

```python
#create a column that says which are disposals
import numpy as np
r21['is_disposal'] = np.where(r21['Stat'] == 'Disposal', 1, np.nan)
```

```python
r21.head(30)
```

```python
#Group by players counting their disposals
Top_disposals = r21.groupby(['Player', 'Team'], as_index = False)['is_disposal'].sum().copy()
Top_disposals.rename(columns = {'is_disposal' : 'Disposals'}, inplace = True)
```

```python
#Limit to those players with at least 400 disposals so far
Top_disposals_400 = Top_disposals[Top_disposals['Disposals'] >= 400].copy()
Top_disposals_400
```

```python
#add a 'Possession' column to determine who has possession and a 'Poss_id' column to give each possession an id

#turning the dataframe into a dictionary makes this happen much faster

r21_dict = r21.to_dict('records')

poss = None
quarter = 0
Poss_list = []
Poss_id_list = []
is_goal_list = []
n=0
for row in r21_dict:
    if row['Quarter'] != quarter:
        poss = None
        Poss_id_list.append(None)
        quarter = row['Quarter']
        row['Possession'] = poss
        Poss_list.append(poss)
        is_goal_list.append(0)
    else:
    
        if row['Stat'] in ['Poss Gain', 'Centre Clearance', 'Stoppage Clearance', 'Kick In', 'Contested Possession']:
            poss = row['Team_status']
            row['Possession'] = poss
            Poss_list.append(poss)
            is_goal_list.append(0)

        elif row['Stat'] in ['Hitout', 'Hitout To Advantage']:
            poss = None
            row['Possession'] = poss
            Poss_list.append(poss)
            is_goal_list.append(0)
            
        elif row['Stat'] == 'Turnover':
            if poss == 'home':
                poss = 'away'
            else:
                poss = 'home'
            row['Possession'] = poss
            Poss_list.append(poss)
            is_goal_list.append(0)
            
        elif row['Stat'] not in ['Hitout', 'Hitout To Advantage', 'Poss Gain', 'Centre Clearance', 'Stoppage Clearance', 'Kick In', 'Contested Possession', 'Goal', 'Behind', 'Turnover']:
            row['Possession'] = poss
            Poss_list.append(poss)
            is_goal_list.append(0)
        
        elif row['Stat'] in ['Goal', 'Behind']:
            row['Possession'] = poss
            Poss_list.append(poss)
            poss = None
            is_goal_list.append(row['Stat'])

#possession counter        
        
        if row['Possession'] == None:
            Poss_id_list.append(None)

        elif (row['Possession'] == Poss_list[-2] and is_goal_list[-2] != 'Goal') or row['Possession'] == 'score':
            Poss_id_list.append(Poss_id_list[-1])
        else:
            n+=1
            Poss_id_list.append(n)


r21 = pd.DataFrame(r21_dict)          
            

r21['Poss_id'] = Poss_id_list

```

```python
r21
```

```python
#Now we go back to the old dataframe and put in a points column based on the stats that are scores.
#Then will group by the possession to get the points scored each possession
r21['Points'] = np.where(r21['Stat'] == 'Goal', 6, np.where(r21['Stat'] == 'Behind', 1, 0))
r21_points = r21.groupby('Poss_id', as_index = False)['Points'].sum().copy()
```

```python
r21_points
```

```python
#Next let's make a dictionary that tells us how many points come from each possession
poss_score_dict = {}
for ind, row in r21_points.iterrows():
    poss_score_dict[row['Poss_id']] = row['Points']
```

```python
poss_score_dict
```

```python
#Now we're going to look at the possessions that each of the disposal leaders had a disposal in and see how many points they resulted in.
#And we'll take an average per possession

#(we'll make a new r21_dict for some wait-reduction and make a TD400 dictionary too)
r21_dict = r21.to_dict('records')
td400_dict = Top_disposals_400.to_dict('records')
team_possessions_list = []
total_points_list = []
avg_points_list = []
for row in td400_dict:
    poss_ids = []
    for arow in r21_dict:
        if row['Player'] == arow['Player'] and row['Team'] == arow['Team'] and arow['is_disposal'] == 1 and arow['Poss_id'] > 0:
            if arow['Poss_id'] not in poss_ids:
                poss_ids.append(arow['Poss_id'])
    team_possessions_list.append(len(poss_ids))
    
    total_points = 0
    for n in poss_ids:
        total_points += poss_score_dict[n]
    total_points_list.append(total_points)
    avg_points_list.append(total_points/len(poss_ids))
    
        
        
Top_disposals_400['Possessions'] = team_possessions_list
Top_disposals_400['Total_points'] = total_points_list
Top_disposals_400['Avg_points'] = avg_points_list
   
    
```

```python

Top_disposals_400 = Top_disposals_400.sort_values('Avg_points', ascending = False).reset_index()
Top_disposals_400.index = Top_disposals_400.index + 1
Top_disposals_400.drop(columns = ['index'], inplace = True)
Top_disposals_400.head(20)
```

## Which players contribute to the highest percentage of their team's goals?

```python
#First we'll make a dictionary with teams a keys and the values are a list of possession numbers they scored a goal in.

Team_goals_dict = {}
r21_dict = r21.to_dict('records')
for row in r21_dict:
    if row['Stat'] == 'Goal':
        if row['Team'] not in Team_goals_dict.keys():
            Team_goals_dict[row['Team']] = [row['Poss_id']]
        else:
            Team_goals_dict[row['Team']].append(row['Poss_id'])
```

```python
Team_goals_dict
```

```python
#Next we'll make a dictionary recording, for each player, which of their team's possessions that led to goals,
# they had a disposal in.
players_d_to_goal = {}
for row in r21_dict:
    if row['Poss_id'] in Team_goals_dict[row['Team']] and row['is_disposal'] == 1:
        if row['Player'] not in players_d_to_goal.keys():
            players_d_to_goal[row['Player']] = [row['Poss_id']]
        else:
            if row['Poss_id'] not in players_d_to_goal[row['Player']]:
                players_d_to_goal[row['Player']].append(row['Poss_id'])
```

```python
players_d_to_goal
```

```python
#Let's put this info in a dataframe.

players_d_to_goal_df = pd.DataFrame()
```

```python
players_d_to_goal_df['Players'] = players_d_to_goal.keys()
```

```python
#how many of their team's goals did they contribute to?
for ind, row in players_d_to_goal_df.iterrows():
    players_d_to_goal_df.loc[ind, 'Disposal to goal'] = len(players_d_to_goal[row['Players']])
```

```python
players_d_to_goal_df
```

```python
#These are just a few steps leading to tuples with players and their teams, so we can add a team column to
#the players_d_to_goal_df.

r21_teams = r21.groupby(['Player', 'Team'], as_index = False)['Stat'].count()
```

```python
r21_teams.head(10)
```

```python
P_T = list(zip(r21_teams['Player'].to_list(), r21_teams['Team'].to_list()))
```

```python
for ind, row in players_d_to_goal_df.iterrows():
    for n in P_T:
        if row['Players'] == n[0]:
            players_d_to_goal_df.loc[ind, 'Team'] = n[1]
```

```python
players_d_to_goal_df
```

```python
#Now we add a column for the total possessions their team had that led to goals
#and a column for the percentage of those goals which they contributed to.

#VOILA!

for ind, row in players_d_to_goal_df.iterrows():
    players_d_to_goal_df.loc[ind, 'Team Goals'] = len(Team_goals_dict[row['Team']])
```

```python
players_d_to_goal_df['Pct of Team Goals'] = players_d_to_goal_df['Disposal to goal']/players_d_to_goal_df['Team Goals']
```

```python
players_d_to_goal_df = players_d_to_goal_df.sort_values('Pct of Team Goals', ascending = False).reset_index()
players_d_to_goal_df.index = players_d_to_goal_df.index + 1
players_d_to_goal_df = players_d_to_goal_df.drop(columns = 'index')
players_d_to_goal_df.head(20)
```
